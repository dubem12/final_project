```python

start_year = 1900
end_year = 2015
range_size = end_year - start_year

# Calculate the approximate number of bins using the square root rule
num_bins = int(np.sqrt(range_size))
```

```python
num_bins
```

```python
#itinerate yr_built column and create a 10 years bin for very value in the column
# Define the range
start_year = 1900
end_year = 2015

# Calculate the range size
range_size = end_year - start_year

# Calculate the bin size
bin_size = range_size // 10

# Create a list to store the bin labels
bin_labels = []

# Assign each year to the appropriate bin label
for year in range(start_year, end_year + 1, bin_size):
    bin_labels.append(f'{year}-{year + bin_size - 1}')

# Create a new column in your DataFrame and assign bin labels based on the year
df['year_bin'] = pd.cut(df['yr_built'], bins=range(start_year, end_year + bin_size, bin_size), labels=bin_labels, include_lowest=True)

# Display the resulting DataFrame
```
